#### **Focusing on time series engineering** !!

```python
train_long_subset.head(10)
```

```python
#Creating basic features
train_long_subset['year'] = train_long_subset['date'].dt.year
train_long_subset['month'] = train_long_subset['date'].dt.month
train_long_subset['day'] = train_long_subset['date'].dt.day
train_long_subset['weeknumber'] = train_long_subset['date'].dt.week
train_long_subset['dayofweek'] = train_long_subset['date'].dt.dayofweek
```

```python
train_long_subset.head()
```

```python
#Creating the lags and moving averages
dayLags = [7, 28]
lagSalesCols = [f"lag_{dayLag}" for dayLag in dayLags]
for dayLag, lagSalesCol in zip(dayLags, lagSalesCols):
    train_long_subset[lagSalesCol] = train_long_subset[["identifier","value"]].groupby("identifier")["value"].shift(dayLag)


    
windows = [7, 28]
for window in windows:
    for dayLag, lagSalesCol in zip(dayLags, lagSalesCols):
        train_long_subset[f"rmean_{dayLag}_{window}"] = train_long_subset[["identifier", lagSalesCol]].groupby("identifier")[lagSalesCol].transform(lambda x: x.rolling(window).mean())
```
